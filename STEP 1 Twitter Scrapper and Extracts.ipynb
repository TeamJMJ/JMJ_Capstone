##### 1. Code for extracting hourly sample data from twitter for 3 years. It takes about 23-24 hours to extract 1 month of data. So we ran 12 paralllel notebooks, each extracing 400 random tweets every hour containg the word bitcoin.

##### Below is an example extracting one month of data for the month of January 2022


```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
from datetime import datetime, date, timedelta
import random
```

## Hourly sampling

```python
%%time
##WORKING SO FAR V4
startday = "2022-01-01 02:00"
date_1 = datetime.strptime(startday,'%Y-%m-%d %H:%M')
day1 = date_1.timestamp()

day = 31
numberoftweets = 420
RandomSample = 0.8
tweets_list2 = []
for i in range(day):
    print(i)
    for j in range(25):
        #print(j)
        x = 0
        for tweet in sntwitter.TwitterSearchScraper('bitcoin since_time:{} until_time:{}'.format(int(day1), int((date_1 + timedelta(days = i, hours=j)).timestamp()))).get_items():
            #print(x)
            if x==numberoftweets:
                break
            if random.random() >RandomSample:  
                tweets_list2.append([tweet.user.username, tweet.rawContent,tweet.date,tweet.likeCount,tweet.retweetCount,tweet.user.followersCount,tweet.user.friendsCount])
                x +=1
                #print(x)

tweets_df2 = pd.DataFrame(tweets_list2, columns=['user', 'tweets','time','favorites','retweets','number_of_followers','following'])
```
