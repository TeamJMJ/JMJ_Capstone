```python
def signal(row):
    if row['Prediction'] >= 0.525:
        return 1
    elif row['Prediction'] < 0.475:
        return -1
    else:
        return 0

prediction_df_list = [all_LSTM_df, all_ensemble_df, all_stacking_1GBC_df, all_stacking_2LR_df, all_stacking_3LSTM_df,priority_LSTM_df, 
                       priority_ensemble_df, priority_stacking_1GBC_df, priority_stacking_2LR_df, 
                      priority_stacking_3LSTM_df]

# Loop through the dataframes and rename columns as necessary
for df in prediction_df_list:
    if 'Predictions' in df.columns:
        df.rename(columns={'Predictions': 'Prediction'}, inplace=True)
        df.reset_index()
    elif 'Predicted_Label' in df.columns:
        df.rename(columns={'Predicted_Label': 'Prediction'}, inplace=True)
        df.reset_index()
for df in prediction_df_list:
    df['Signal'] = df.apply(signal, axis=1)
    
```

```python
def backtest_strategy(df, starting_investment):
    # Add Aggregate Signal column
    df['Aggregate Signal'] = df['Signal'] #(df['Signal'].shift(1) + df['Signal'])
    df['Aggregate Signal'][0] = df['Signal'][0]
    
    # Set starting values
    cash_balance = starting_investment/2
    shares = cash_balance / df.iloc[0]['Close']
    invested_balance = starting_investment

    # Create backtest columns
    df["Cash Balance"] = cash_balance
    df["Shares"] = shares
    df["Invested Balance"] = invested_balance
    
    # Calculate initial buy and hold shares
    initial_share_price = df['Close'][0]
    buy_and_hold_shares = starting_investment / initial_share_price
    
    # Calculate buy and hold balance for each row
    df['Buy and Hold Shares'] = buy_and_hold_shares
    df['Buy and Hold Balance'] = df['Close'] * df['Buy and Hold Shares']

    # Loop through rows in dataframe
    for i, row in df.iterrows():
        if i == 0:
            continue

        # Calculate Invested Balance
        invested_balance = df.loc[i-1,'Cash Balance'] + (df.loc[i-1,'Shares'] * row['Close'])
        
        # Calculate Cash Balance
        if row['Aggregate Signal'] == 0:
            cash_balance = df.loc[i-1,'Cash Balance']
        #elif row['Aggregate Signal'] in [1, 2]:
            #cash_balance = invested_balance / 2
        elif row['Aggregate Signal'] in [1, 2]:
            cash_balance = 0
        #elif row['Aggregate Signal'] in [-1, -2]:
            #cash_balance = shares/2 * row['Close'] + df.loc[i-1,'Cash Balance']
        elif row['Aggregate Signal'] in [-1, -2]:
            cash_balance = invested_balance

        # Calculate Shares
        if row['Aggregate Signal'] == 0:
            shares = df.loc[i-1,'Shares']
        #elif row['Aggregate Signal'] in [1, 2]:
            #shares = cash_balance / row['Close']
        elif row['Aggregate Signal'] in [1, 2]:
            shares = invested_balance / row['Close']
        #elif row['Aggregate Signal'] in [-1, -2]:
            #shares = df.loc[i-1,'Shares']/2
        elif row['Aggregate Signal'] in [-1, -2]:
            shares = 0

        # Update dataframe with new values
        df.loc[i, 'Cash Balance'] = cash_balance
        df.loc[i, 'Shares'] = shares
        df.loc[i, 'Invested Balance'] = invested_balance
    
    return df
```
